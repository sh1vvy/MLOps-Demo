```python
arr2 = [(x,data[x].isnull().mean()*100) for x in data]
arr3 = pd.DataFrame(arr2, columns=['column_name', 'percentages'])
arr3
```

```python
drope = list(arr3[arr3['percentages']>50]['column_name'])
drope
```

```python
data.drop(drope, axis=1, inplace = True)
```

```python
unique = data.nunique()
unique=unique[unique.values==1]
data.drop(labels=list(unique.index), axis = 1, inplace = True)
```

```python
drope_col = ["id", "member_id","url", "zip_code"]
data.drop(labels= drope_col, axis=1, inplace = True)
print(data.shape, "rows & columns .")
```

```python
data['int_rate']=(data['int_rate'].str.strip("%").astype(float))
```

```python
num_cols =['loan_amnt', 'funded_amnt', 'funded_amnt_inv', 'installment', 'int_rate','dti']
data[num_cols]=data[num_cols].apply(pd.to_numeric)
```
